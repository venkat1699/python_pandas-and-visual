### **Visualizations**

```python
import matplotlib.pyplot as plt
import seaborn as sn
import pandas as pd
```

```python
df1= pd.read_csv("Universities.csv")
```

```python
df1
```

```python
plt.hist(df1['Accept'])
plt.title('Histogram of Accept')
plt.xlabel('Accept')
plt.ylabel('Frequency')
plt.show()
```

```python
sn.histplot(df1['Accept'])
plt.show()
```

```python
sn.histplot(df1['SAT'])
plt.show()
```

```python
plt.plot(df1['SAT'],df1['Accept'],'rs')
plt.title('Scatter plot btw SAT vs Accept')
plt.xlabel('SAT')
plt.ylabel('Accept')
plt.show()
```

```python
sn.scatterplot(x='SAT',y='Accept',data=df1)
plt.title('Scatter plot btw SAT vs Accept')
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")
sn.distplot(df1['Accept'])
plt.show()
```

```python
sn.distplot(df1['Top10'])
plt.show()
```

```python
sn.pairplot(df1)
plt.show()
```

```python
sn.pairplot(df1.iloc[:,1:4])
plt.show()
```

```python
plt.boxplot(df1['SAT'], vert = False)
plt.show()
```
